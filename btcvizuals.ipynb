```python
%reload_ext blackcellmagic
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker 
import seaborn as sns
import os
```

```python
path = os.path.join(
    "Bootcamp", "HWdir", "CryptoReport", "Resources", "bitcoinity_data.csv"
)
bitcoinity_df = pd.read_csv(path).set_index("Time")
dates = []
for t in bitcoinity_df.index:
    dates.append(t[0:10])
bitcoinity_df.insert(0, "date", dates)
```

```python
early_2011 = bitcoinity_df.loc["2011-01-03 00:00:00 UTC":"2011-06-20 00:00:00 UTC"]
##faulty data after 6/20/2011

x_ticks = ["Jan'11", "Feb'11", "Mar'11", "Apr'11", "May'11", "June'11"]

sns.set(style="dark", font_scale=1.8, rc={"lines.linewidth": 3})

fig, ax1 = plt.subplots(figsize=(12, 12))
plt.title("Bitcoin Price/Volume in 2011", fontsize=30)
ax2 = ax1.twinx()

sns.barplot(x="date", y="volume", color="b", data=early_2011, ax=ax1)

sns.lineplot(x="date", y="price", color="r", data=early_2011, ax=ax2)
ax1.yaxis.set_major_formatter(ticker.EngFormatter())
ax1.set_xlabel("")
ax1.set_ylabel("Volume  (USD)")
ax2.set_ylabel("Price  (USD)")
plt.xticks(np.arange(0, len(early_2011), step=30), x_ticks)
plt.savefig("early_2011.jpg")
plt.show()
```
